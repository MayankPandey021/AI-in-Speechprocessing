```python
## Loading the audio file
import librosa
import matplotlib.pyplot as plt
speech = r'C:\Users\HPELITE\Desktop\mayank_speech.wav'
y, sr = librosa.load(speech)
## plotting
librosa.display.waveshow(y)
```

```python
## Observing the length & magnitude range

print("Finding the Length and the Magnitude of the signal:")
print(f"Signal Length: {len(y)} samples")
print(f"Magnitude Range: {y.min()} to {y.max()}")
```

```python
## Playing signal

from pydub import AudioSegment
from pydub.playback import play
audio = AudioSegment(data=y.tobytes(), sample_width=y.dtype.itemsize, frame_rate=sr, channels=1)
play(audio)
```

```python
## part of the signal 

import sounddevice as sd
import librosa
speech = r'C:\Users\HPELITE\Desktop\mayank_speech.wav'
y, sr = librosa.load(speech)
starting = 0
finishing = 2
segment = y[int(starting * sr):int(finishing * sr)]
sd.play(segment, samplerate=sr)
sd.wait()
```

```python
## segmenting the signal into various components and playing the signal

import IPython.display as ipd
num_segments = 3
segment_duration = len(y) // num_segments
print("the various segments")
for i in range(num_segments):
    start_sample = i * segment_duration
    end_sample = (i + 1) * segment_duration
    segment = y[start_sample:end_sample]
    ipd.display(ipd.Audio(segment, rate=sr))
```
